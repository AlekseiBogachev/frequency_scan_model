```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
anscombe = sns.load_dataset('anscombe')
anscombe.rename(columns = {'dataset': 'Набор данных'}, inplace=True)
anscombe.head(10)
```

```python
sns.set(font_scale = 2)
sns.set_style("whitegrid")
```

```python
sns.lmplot(x="x", y="y", col='Набор данных', data=anscombe, col_wrap=2, ci=None, scatter_kws={"s": 80})
plt.savefig('../images/anscombe_quartet.pdf', bbox_inches='tight')
plt.show()
```

```python
predict = lambda x: 3 + 0.5 * x
from sklearn.metrics import mean_squared_error
anscombe.rename(columns = {'Набор данных': 'dataset'}, inplace=True);
```

```python
subset = anscombe.query("dataset == 'I'")
X_train = subset.x.to_numpy()
y_train = subset.y.to_numpy()

display(subset)

print(f'RMSE = {mean_squared_error(y_train, predict(X_train), squared=False)}')
```
