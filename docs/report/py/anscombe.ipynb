```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
anscombe = sns.load_dataset('anscombe')
anscombe.head(10)
```

```python
sns.set_style("whitegrid")
```

```python
sns.lmplot(x="x", y="y", col='dataset', data=anscombe, col_wrap=2, ci=None, scatter_kws={"s": 80})
plt.savefig('../images/anscombe_quartet.pdf', bbox_inches='tight')
plt.show()
```
