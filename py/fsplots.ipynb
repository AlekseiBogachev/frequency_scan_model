```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = ''
```

```python
import numpy as np
import matplotlib.pyplot as plt
from fsplots import plot_spectr
from fsplots import plot_model
from fsplots import plot_loss_path
from fsplots import plot_deviations
from fsplots import plot_experimental_points

from fsmodels import SklMultiExpFrequencyScan, SklSingleExpFrequencyScan

%matplotlib inline
```

```python
exps_params = np.array([[-0.5, 2],
                        [-1, 1],
                        [-2, 2],
                        [-3, 1]
                       ])
```

# plot_spectr

```python
plot_spectr(exps_params, ylim=[0,3])
plt.show()
```

# plot_model

```python
X = np.linspace(0, 3.5, 1000)

fs = SklSingleExpFrequencyScan(filling_pulse = 20*10**-6,
                               fit_p_coef = True,
                               learning_rate = 0.1,
                               n_iters = 200, 
                               stop_val = 10**-5,
                               verbose = True)

fs.exps_params_ = [[-2.5, 3.5]]
fs.p_coef_ = 0.3

y = fs.predict(X)
y += np.random.normal(loc=0, scale=0.2, size=y.shape[0])

fs.fit(X, y)
```

```python
plot_model(X, y, SklSingleExpFrequencyScan, fs.fit_results_)
plt.show()
```

```python
X = np.linspace(0, 3.5, 1000)

fs = SklMultiExpFrequencyScan(filling_pulse = 20*10**-6,
                              n_exps = 3,
                              learning_rate = 0.1,
                              n_iters = 200, 
                              stop_val = 10**-5,
                              verbose = True)

fs.exps_params_ = [[-1.0, 3.5],
                   [-2.3, 3.0],
                   [-3.5, 2.5]]

y = fs.predict(X)
y += np.random.normal(loc=0, scale=0.2, size=y.shape[0])

fs.fit(X, y)
```

```python
plot_model(X, y, SklMultiExpFrequencyScan, fs.fit_results_)
plt.show()
```

# plot_loss_path

```python
plot_loss_path(fs.fit_results_)
plt.show()
```

# plot_deviations

```python
plot_deviations(X, y, fs.predict(X))
plt.show()
```

# plot_experimental_points

```python
plot_experimental_points(X, y)
plt.show()
```
