## Подготовка исходных данных

```python
n_points = 30
```

```python
filling_pulse = np.linspace(0, 40*10**-6, n_points)
```

```python
tau_pow = np.linspace(-3, 2, n_points)
```

```python
def get_M(filling_pulse, tau_pow):
    t_pow = tf.Variable(tau_pow, dtype='float64')
    model = BaseModel(filling_pulse=filling_pulse)
    return model._get_M(time_constant_power=t_pow,
                        learning_rate=0.2, 
                        n_iters=100, 
                        stop_val = 10**-10)
```

```python
filling_pulse, tau_pow = np.meshgrid(filling_pulse, tau_pow)

X = filling_pulse.flatten()
Y = tau_pow.flatten()

Z = Parallel(n_jobs=-1)(delayed(get_M)(x, y) for x, y in zip(X, Y))
```

```python
Z = np.reshape(np.c_[Z], filling_pulse.shape)
M = Z
```

```python
fig, ax = plt.subplots(figsize=(10,10))

CS = ax.contour(filling_pulse, tau_pow, M,
#                 levels=[0, 0.05, 0.1, 0.25, 0.5, 1, 2.0, 3.0, 4.0, 5.0], 
                colors='k')

ax.clabel(CS,
          inline=True, 
          fontsize=12)

IM = ax.imshow(M,
#                interpolation='bilinear',
               origin='lower',
               cmap=cm.PuBuGn)

cbar = fig.colorbar(IM, shrink=1)
cbar.ax.set_ylabel('M')

plt.grid()

plt.show()
```
