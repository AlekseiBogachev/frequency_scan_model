```python
import matplotlib.pyplot as plt
import numpy as np

from fsmisc import plot_spectr
from fsmisc import plot_model

from fsmodels import SklMultiExpFrequencyScan, SklSingleExpFrequencyScan
```

```python
X = np.linspace(0, 3.5, 100)

exps_params = np.array([[-0.5, 2],
                        [-1, 1],
                        [-2, 2],
                        [-3, 1]
                       ])
```

# plot_spectr

```python
plot_spectr(exps_params, ylim=[0,3])
plt.show()
```

# plot_model

```python
X = np.linspace(0, 3.5, 100)

fs = SklSingleExpFrequencyScan(filling_pulse = 20*10**-6,
                               fit_p_coef = True,
                               learning_rate = 0.1,
                               n_iters = 200, 
                               stop_val = 10**-5,
                               verbose = True)

fs.exps_params_ = [[-2.5, 3.5]]
fs.p_coef_ = 0.3

y = fs.predict(X)
y += np.random.normal(loc=0, scale=0.2, size=y.shape[0])

fs.fit(X, y)
```

```python
plot_model(X, y, SklSingleExpFrequencyScan, fs.fit_results_)
plt.show()
```

```python
X = np.linspace(0, 3.5, 100)

fs = SklMultiExpFrequencyScan(filling_pulse = 20*10**-6,
                              n_exps = 3,
                              learning_rate = 0.1,
                              n_iters = 200, 
                              stop_val = 10**-5,
                              verbose = True)

fs.exps_params_ = [[-1.0, 3.5],
                   [-2.3, 3.0],
                   [-3.5, 2.5]]

y = fs.predict(X)
y += np.random.normal(loc=0, scale=0.2, size=y.shape[0])

fs.fit(X, y)
```

```python
plot_model(X, y, SklMultiExpFrequencyScan, fs.fit_results_)
plt.show()
```
