```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

from pydlts.fsmodels import SklMultiExpFrequencyScan
from pydlts.fsplots import plot_experimental_points
from pydlts.fsplots import plot_model
```

# Проверка прототипа модели на частотном скане из 3 экспоненциальных составляющих

```python
f_pulse = 20 * 10**-6
fs = SklMultiExpFrequencyScan(n_exps=3, filling_pulse = f_pulse)

fs.exps_params_=[[-1.0, 3.5],
                 [-2.3, 3.0],
                 [-3.5, 2.5]]
```

```python
X = np.linspace(0, 3.5, 200)
y = fs.predict(X) 
y += np.random.normal(loc=0, scale=0.2, size=y.shape[0])
```

```python
plot_experimental_points(X, y)
plt.show()
```

```python
n_exps = 3

initial_exps_params_ = [[-1.5, 1.5],
                        [-2.0, 1.0],
                        [-3.0, 1.0]
                       ]

fs = SklMultiExpFrequencyScan(filling_pulse = f_pulse,                           
                              n_exps=n_exps,
                              n_iters=200,
                              stop_val=10**-5,
                              verbose=True
                             )

fs.fit(X, y, initial_exps_params_=initial_exps_params_)
```

```python
plot_model(X, y, SklMultiExpFrequencyScan, fs.fit_results_)
plt.show()
```

```python
fit_results = fs.fit_results_
fit_results
```

```python
fs.exps_params_
```

# Проверка прототипа модели на частотном скане из 5 экспоненциальных составляющих

```python
fs = SklMultiExpFrequencyScan(n_exps=5, filling_pulse = f_pulse)

fs.exps_params_=[[-1.0, 3.50],
                 [-1.7, 3.25],
                 [-2.3, 3.00],
                 [-2.9, 2.75],
                 [-3.5, 2.50]]

X = np.linspace(0, 3.5, 200)
y = fs.predict(X) 
y += np.random.normal(loc=0, scale=0.2, size=y.shape[0])

plot_experimental_points(X, y)
plt.show()
```

```python
n_exps = 5

fs = SklMultiExpFrequencyScan(filling_pulse = f_pulse,                           
                              n_exps=n_exps,
                              n_iters=200,
                              stop_val=10**-5,
                              verbose=True
                             )

fs.fit(X, y)
```

```python
plot_model(X, y, SklMultiExpFrequencyScan, fs.fit_results_)
plt.show()
```

```python
fit_results = fs.fit_results_
fit_results
```

```python
fs.exps_params_
```

# Пример неудачной идентификации - недостаточное количество экспоененциальных составляющих

```python
plot_experimental_points(X, y)
plt.show()
```

```python
n_exps = 3

fs = SklMultiExpFrequencyScan(filling_pulse = f_pulse,                           
                              n_exps=n_exps,
                              n_iters=200,
                              learning_rate=0.075,
                              stop_val=10**-5,
                              verbose=True
                             )

fs.fit(X, y)

plot_model(X, y, SklMultiExpFrequencyScan, fs.fit_results_)
plt.show()
```
